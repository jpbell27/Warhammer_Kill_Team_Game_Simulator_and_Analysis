```python
def operative_attrition(operatives):
    # Define the probability range (in percentage)
    prob_range = (0, operatives)

    # Define the number to potentially reduce
    number = operatives

    # Define the scaling factor for the probability range
    scaling_factor = 0.125

    # Calculate the probability range based on the size of the number
    scaled_prob_range = (
        prob_range[0] + scaling_factor * number,
        prob_range[1] + scaling_factor * number
    )

    # Generate a random number between 0 and the number of operatives
    random_number = random.randint(0, operatives)

    # Check if the random number falls within the probability range
    if random_number >= scaled_prob_range[0] and random_number <= scaled_prob_range[1]:
        # Generate a random factor between 0.5 and 0.9
        reduction_factor = random.uniform(0.5, 0.9)
        # Reduce the number by the factor
        number *= reduction_factor

    # Ensure that the number is at least 1
    number = max(1, round(number))

    # Return the result instead of printing it
    return number
```
